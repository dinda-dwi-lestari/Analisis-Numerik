# **STUDI KASUS KELOMPOK 1**
Data berikut menunjukkan rata-rata tinggi muka air sungai (cm) pada salah satu wilayah rawa gambut yang terhubung dengan Daerah Aliran Sungai (DAS) selama beberapa bulan pengamatan.

Variabel :


*   Variabel independen (x) : waktu (bulan)
*   Variabel dependen (y) : tinggi muka air sungai (cm)







```
Bulan (x)	Tinggi Muka Air (y)
 1          120
 2	        135
 3	        150
 4	        170
 5	        195
 6	        215
```

Tentukan interpolasi untuk mengestimasi tinggi muka air pada bulan ke-4,5.









## **Perhitungan Manual**


### Interpolasi Linear

Diketahui:

\begin{align*}
x &= 4.5 \\
x_1 &= 4,\quad y_1 = 170 \\
x_2 &= 6,\quad y_2 = 215
\end{align*}

Rumus interpolasi linear:

$$
y = \frac{x - x_1}{x_2 - x_1}(y_2 - y_1) + y_1
$$

Substitusi:

$$
y = \frac{4.5 - 4}{6 - 4}(215 - 170) + 170
$$

$$
y = \frac{0.5}{2} \times 45 + 170
$$

$$
y = 0.25 \times 45 + 170
$$

$$
y = 11.25 + 170
$$

$$
y = 181.25
$$





### Interpolasi Lagrange Orde 2

Diketahui:

Diketahui:
\begin{align*}
x &= 4.5 \\
x_0 &= 6,\quad y_0 = 215 \\
x_1 &= 5,\quad y_1 = 195 \\
x_2 &= 4,\quad y_2 = 170
\end{align*}

Rumus:

$$
P_2(x) =
\frac{(x - x_1)(x - x_2)}{(x_0 - x_1)(x_0 - x_2)} y_0 +
\frac{(x - x_0)(x - x_2)}{(x_1 - x_0)(x_1 - x_2)} y_1 +
\frac{(x - x_0)(x - x_1)}{(x_2 - x_0)(x_2 - x_1)} y_2
$$

Substitusi:

$$
P_2 =
\frac{(4.5 - 5)(4.5 - 4)}{(6 - 5)(6 - 4)} (215) +
\frac{(4.5 - 6)(4.5 - 4)}{(5 - 6)(5 - 4)} (195) +
\frac{(4.5 - 6)(4.5 - 5)}{(4 - 6)(4 - 5)} (170)
$$

$$
P_2 =
-0.125(215) + 0.75(195) + 0.375(170)
$$

$$
P_2 = -26.875 + 146.25 + 63.75
$$

$$
P_2 = 183.125
$$


## **Perhitungan Phyton**

### Interpolasi Linear

````python
import numpy as np

# Data titik
x = np.array([4, 6], dtype=float)
y = np.array([170, 215], dtype=float)

# Titik yang ingin dicari
x_target = 4.5


def interpolasi_linear(x, y, x_target):

    x0, x1 = x[0], x[1]
    y0, y1 = y[0], y[1]

    print("=== Interpolasi Linear ===")
    print(f"Titik: ({x0},{y0}) dan ({x1},{y1})")

    print("\nRumus:")
    print(f"f(x) =(({x_target}-{x0})/({x1}-{x0})) * ({y1}-{y0}) + {y0}")

    # perhitungan
    hasil = ((x_target - x0)/(x1 - x0)) * (y1 - y0) + y0

    return hasil


hasil = interpolasi_linear(x, y, x_target)

print("\nHasil interpolasi linear:", hasil)
````

### Interpolasi Langrange Orde 2

````python
import numpy as np

# Data titik
x = np.array([6,5,4], dtype=float)
y = np.array([215,195,170], dtype=float)

# Titik yang ingin dicari
x_target = 4.5

def lagrange_interpolation(x, y, x_target):
    n = len(x)
    result = 0

    for i in range(n):
        term = y[i]
        print(f"\n=== Suku ke-{i} ===")

        for j in range(n):
            if j != i:
                print(f"(({x_target} - {x[j]}) / ({x[i]} - {x[j]}))")
                term *= (x_target - x[j]) / (x[i] - x[j])

        print(f"Nilai suku ke-{i}: {term}")
        result += term

    return result

hasil = lagrange_interpolation(x, y, x_target)
print("\nHasil interpolasi:", hasil)
````

## **Interpretasi**

Berdasarkan hasil perhitungan:


*   Interpolasi Linear menghasilkan nilai 181,25

*   Interpolasi Lagrange Orde 2 menghasilkan nilai 183,125




Perbedaan hasil ini terjadi karena jumlah titik yang digunakan berbeda. **Interpolasi linear** hanya menggunakan **2 titik**, sehingga hasilnya cenderung lebih sederhana dan kurang menangkap perubahan data secara detail. Sedangkan **interpolasi Lagrange orde 2** menggunakan **3 titik**, sehingga mampu mengikuti pola data dengan lebih baik.

Dari data yang diberikan, terlihat bahwa nilai tinggi muka air mengalami peningkatan yang tidak sepenuhnya linear. Oleh karena itu, hasil dari Lagrange orde 2 lebih mendekati kondisi sebenarnya dibandingkan interpolasi linear.

**Interpolasi Lagrange orde 2 memberikan hasil yang lebih akurat dibandingkan interpolasi linear karena mempertimbangkan lebih banyak titik data**.

**Tinggi muka air pada bulan ke-4,5 adalah 183,125 cm**
